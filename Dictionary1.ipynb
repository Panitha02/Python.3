```python
#Dictionary Methods 
d={1:2,2:3} 
d1={1:2,2:3} 
print(id(d),id(d1))#dictionary is mutable 
d={"manikanta":23456,"narendar":765432,"saisidh":65432,"ganesh":876543} 
print(type(d)) 
print(d["manikanta"]) 
print(d["narendar"]) 
print(d.get("manikanta"))#variable.get(key) 
print(d.get("pardhu"))#it will not return any error when key is not present 
d.update({"manikanta":34567,"a":765432})#it updates key when key is present in dictionary otherwise it add key and value to dictionary 
print(d) 
d.pop("manikanta") 
print(d)# it removes value of particular key and returns dictionary after removing particular key 
d.popitem()# it removes last key and value pair from dictionary 
print(d) 
print(d.keys())# it returns list of keys in dictionary 
print(d.values())#it returns list of values in dictionary 
print(d.items())#it returns key and values in tuple those tuples are stored in list 
d.setdefault("john",55555555555)#adds key if only doesnt exist 
print(d) 
c=d.copy()#it siimply copies the dictionary 
print(c)
```
