```python
n=4
for i in range(n-1,0,-1):
    for j in range(n,0,-1):
        print(j,end=' ')
    print()
```

```python
for i in range(1,4+1):
    for j in range(4-i):#no.of.spaces
        print(" ",end=' ')
    for k in range(1,i+1):#no.of.numbers(range of numbers)
        print(k,end=' ')
    print()
```
