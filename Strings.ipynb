```python
#String Methods 
a="Panitha is good" 
print(a.upper()) 
print(a.lower()) 
print(a.capitalize()) 
print(a.title()) 
b="PANITHAA" 
print(b.swapcase())
```

```python
#Character Type checking methods 
a="abcd" 
print(a.isalpha()) 
print(a.isdigit()) 
b="123" 
print(b.isdigit() )#returns true if the given string is in digit or superscript (power of number)(number^2 here 2 is superscript) 
c="2^2" 
print(b.isdigit()) 
d="1234" 
print(d.isnumeric())#it returns true if tHe given string is digits or fractions or subscripts(number below the value)(a1 here 1 is subscript)or roman 
d="0" 
print(d.isdigit()) 
a=" " 
print(a.isspace()) 
a="\t" 
print(a.isprintable()) 
a="123abd" 
print(a.isalnum()) 
a="title" 
print(a.islower()) 
a="Lalithya" 
print(a.isupper()) 
a="Lalithya Is Good" 
print(a.istitle())
```

```python
#searching and checking methods 
a="apple" 
print(a.find("p"))#it returns first occurence value 
print(a.find("o"))#it returns -1 when the value is not present 
print(a.rfind("p"))#it returns last index value 
print(a.rfind("q")) 
print(a.index("a"))#it returns index of value 
print(a.count("p"))#it returns count 
print(a.startswith("a"))#it returns true if it starts with particular value
print(a.endswith("e"))#it returns true if it ends with particular value 
```

```python
#Modification Methods 
a=" Panitha " 
print(a.strip()) 
print(a.lstrip()) 
print(a.rstrip()) 
print(a.replace("lith","s")) 
print(a.replace("y",""))#we can use replace to remove particular character in string
```

```python
#splitting and joining methods 
a="panitha, is" 
print(a.split())#it trys to split at spaces 
print(a.split(","))#it trys to split at commas 
print('-'.join(a))#it trys to join with - 
print('*'.join(a))
```

```python
#Advance Formatting 
a='6ls' 
print(a.zfill(5))#it adds leading 0 to string to match width 
print(a.ljust(5,"*"))#it moves string to left and add operator to match width 
print(a.rjust(2,"*"))#it moves string to right and add operator to matchh width 
print(a.center(4,"*")) 
print(a.partition("l"))#it separats the string in to three parts before and after separator 
a="StraßĦe" 
print(a.casefold())#it converts unicode characters to lower case letters     
#note :press windows +r to open character map 
```
