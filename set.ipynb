```python
#Set Methods 
s={1} 
print(s) 
s={1,True} 
print(s) 
s={False,0} 
print(s) 
s={1,2,3,4,True,"str",False}#it takes true=1 and so it returns {1,2,3,4,str,false] 
print(s) 
s.add(0)#it will not add o because set dont allow duplicate values 
print(s) 
s.update({False,True,100}) 
print(s) 
print(s.pop())#it will remove arbitary element since set is unordered 
print(s) 
s.discard("str")#it removes if element present if not it will not arise any error 
print(s) 
s.clear()#it removes entire element in set 
print(s) 
s={1,2,3} 
s.update([4,5])#it returns entire elements as set because list is mutable 
print(s)
```

```python
s=[1,2,3] 
s.append({4,5,6})#it returns [1,2,3{4,5,6}]since set is immutable 
print(s) 
s.extend({99,10})#it returns all elements as set in extend 
print(s) 
s={1,2,3,(4,5,6)} 
print(s) 
s=set(map(int,input().split())) 
print(s) 
print(min(s)) 
print(max(s)) 
print(sum(s))
```
