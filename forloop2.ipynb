```python
#print the factorial of given number
n=int(input())
fact=1
for i in range(n):
    k=n-i
    fact*=k
print(fact)
#or
n=int(input())
fact=1
i=0
while i<n:
    k=n-i
    fact*=k
    i+=1
print(fact)
```
